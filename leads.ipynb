```python
import json
import numpy as np
import pandas as pd 
import seaborn as sns
from commands import COMMANDS
from matplotlib import pyplot as plt

sns.set_style('whitegrid')
```

```python
path = 'data/ad_music_proccessed.csv'
data = pd.read_csv(path,index_col=0)
data.head()
```

```python
gp = data.groupby(data.cookie)['user_id'].apply(lambda x: x.unique().tolist()).reset_index()
gp.columns = ['cookie', 'unique_user_ids']
gp['num_users'] = gp.unique_user_ids.apply(lambda x: len(x) if x != [np.nan] else 0)
sus_cks = gp[gp.num_users>1].sort_values(by='num_users',ascending=False)
sus_cks
```

```python
sus_cks_subscription_status = data[data.cookie.isin(sus_cks.cookie.values)]
sus_cks_subscription_status = pd.crosstab(sus_cks_subscription_status['cookie'],sus_cks_subscription_status['subscription_status']).reset_index()
```

```python
sus_cmds = data[data.cookie.isin(sus_cks.cookie)]
sus_cmds = pd.crosstab(sus_cmds['cookie'], sus_cmds['command']).reset_index()
sus_cmds = sus_cmds.loc[:, (sus_cmds != 0).any(axis=0)]
sus_cmds.head()
```

```python
sus_aritsts = data[data.cookie.isin(sus_cks.cookie)].groupby(data.cookie)['target_artist'].apply(lambda x: x.unique().tolist()).reset_index()
sus_messages = data[data.cookie.isin(sus_cks.cookie)].groupby(data.cookie)['target_user'].apply(lambda x: x.unique().tolist()).reset_index()
sus_messages['num_target_users'] = sus_messages['target_user'].apply(lambda x: len(x) if x != [np.nan] else 0)
leads =  pd.merge(sus_cks, sus_aritsts, on='cookie', how='inner')
leads =  pd.merge(leads, sus_messages, on='cookie', how='inner')
leads =  pd.merge(leads, sus_cmds, on='cookie', how='inner')
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  
     display(leads)
```

```python
leads.to_csv('data/leads.csv')
```
